```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas  as pd
```

```python
from sklearn.datasets import make_moons
```

```python
n_samples = 2000
X,y = make_moons(n_samples,noise=0.03,random_state=42)
```

```python
# create a model

tf.random.set_seed(42) # set random seed

# the model has two layers with 100 and 10 neurons respectively and an output layer
model_1 = tf.keras.Sequential([tf.keras.layers.Dense(100),
                              tf.keras.layers.Dense(10),
                              tf.keras.layers.Dense(1)])

# compile the model
model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.SGD(),
               metrics=["accuracy"])
# fit the model

history_1 = model_1.fit(X,y,epochs=100)
```

```python
model_1.evaluate(X,y)
```

```python
# create model 2
# i am changing the hyperparameters

tf.random.set_seed(42)

model_2 = tf.keras.Sequential([tf.keras.layers.Dense(100),
                              tf.keras.layers.Dense(10),
                              tf.keras.layers.Dense(1)])

model_2.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(),
               metrics=["accuracy"])

history_2=model_2.fit(X,y,epochs=20)
```

```python
# ! Wow what an improvement change the optimizer from SGD to Adam our accuracy improves
# lets plot a loss curve to visualize

pd.DataFrame(history_2.history).plot();
```

```python
def plot_decision_boundary(model,X,y):
    """
      Plots the decision boundary created by a model predicting on X.
      This function has been adapted from two phenomenal resources:
       1. CS231n - https://cs231n.github.io/neural-networks-case-study/
       2. Made with ML basics - https://github.com/GokuMohandas/MadeWithML/blob/main/notebooks/08_Neural_Networks.ipynb
    """
    
   
    x_min,x_max = X[:,0].min()-0.1, X[:,0].max()+0.1
    y_min,y_max = X[:,1].min()-0.1, X[:,1].max()+0.1
    
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,100),
                       np.linspace(y_min,y_max,100))
    
    x_in = np.c_[xx.ravel(),yy.ravel()]
    
    y_pred = model.predict(x_in)
    
    # check for multiclass
    if model.output_shape[-1]>1:
        print("doing multiclass classification")
        y_pred = np.argmax(y_pred,axis=1).reshape(xx.shape)
    else:
        print("doing binary classification")
        y_pred = np.round(np.max(y_pred,axis=1)).reshape(xx.shape)
        
    # plot decision boundary
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

```python
plot_decision_boundary(model_2,X,y)
```

```python
# non-linearity

tf.random.set_seed(42)

model_3 = tf.keras.Sequential([tf.keras.layers.Dense(4,activation = "relu"),
                              tf.keras.layers.Dense(4,activation = "relu"),
                              tf.keras.layers.Dense(1,activation = "sigmoid")])

model_3.compile(loss="binary_crossentropy",optimizer="Adam",
               metrics="accuracy")

history_3 = model_3.fit(X,y,epochs=100)
```

```python
plot_decision_boundary(model_3,X,y)
```

```python
# beautiful
# our model is able to separate the different colours from the other in a binary classification problem

pd.DataFrame(history_3.history).plot()
```
